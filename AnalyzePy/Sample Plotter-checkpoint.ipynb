```python
def detect_local_minima(arr):
    # https://stackoverflow.com/questions/3684484/peak-detection-in-a-2d-array/3689710#3689710
    """
    Takes an array and detects the troughs using the local maximum filter.
    Returns a boolean mask of the troughs (i.e. 1 when
    the pixel's value is the neighborhood maximum, 0 otherwise)
    """
    # define an connected neighborhood
    # http://www.scipy.org/doc/api_docs/SciPy.ndimage.morphology.html#generate_binary_structure
    neighborhood = morphology.generate_binary_structure(len(arr.shape),2)
    # apply the local minimum filter; all locations of minimum value 
    # in their neighborhood are set to 1
    # http://www.scipy.org/doc/api_docs/SciPy.ndimage.filters.html#minimum_filter
    local_min = (filters.minimum_filter(arr, footprint=neighborhood)==arr)
    # local_min is a mask that contains the peaks we are 
    # looking for, but also the background.
    # In order to isolate the peaks we must remove the background from the mask.
    # 
    # we create the mask of the background
    background = (arr==0)
    # 
    # a little technicality: we must erode the background in order to 
    # successfully subtract it from local_min, otherwise a line will 
    # appear along the background border (artifact of the local minimum filter)
    # http://www.scipy.org/doc/api_docs/SciPy.ndimage.morphology.html#binary_erosion
    eroded_background = morphology.binary_erosion(
        background, structure=neighborhood, border_value=1)
    # 
    # we obtain the final mask, containing only peaks, 
    # by removing the background from the local_min mask
    detected_minima = local_min ^ eroded_background
    return np.where(detected_minima) 
```

```python
def cut_measure(graph, n1=0, n2=None):
    if n2 == None:
        n2 = graph.shape[0]
        
    if n1 * n2 >= 0:
        n1, n2 = min(n1, n2), max(n1, n2)
    else:
        l = graph.shape[0]
        if n1 < 0:
            n1, n2 = min(l + n1, n2), max(l + n1, n2)
        else:
            n1, n2 = min(n1, l + n2), max(n1, l + n2)
    
    A = graph[0:n1, 0:n1].sum() + graph[0:n1, 0:n1].trace()
    B = graph[n1:n2, 0:n1].sum()
    C = graph[n2:, 0:n1].sum()
    
    D = graph[0:n1, n1:n2].sum()
    E = graph[n1:n2, n1:n2].sum() + graph[n1:n2, n1:n2].trace()
    F = graph[n2:, n1:n2].sum()
    
    G = graph[0:n1, n2:].sum()
    H = graph[n1:n2, n2:].sum()
    I = graph[n2:, n2:].sum() + graph[n2:, n2:].trace()
    
    cut1 = B + C + D + G
    assov1 = cut1 + A
    m1 = 1 if assov1 == 0 else cut1 / assov1
    
    cut2 = B + F + H + D
    assov2 = cut2 + E
    m2 = 1 if assov2 == 0 else cut2 / assov2
    
    cut3 = C + F + G + H
    assov3 = cut3 + I
    m3 = 1 if assov3 == 0 else cut3 / assov3
    
    return m1 + m2 + m3
```

```python
def method1(graph, pv):
    for k, graph in pv.items():
        Cab = np.zeros((graph.shape[0], ))
        Aav = np.zeros((graph.shape[0], ))
        Abv = np.zeros((graph.shape[0], ))
        for n in range(graph.shape[0]):
            Cab[n] = graph[:n, n:].sum()
            Aav[n] = graph[:n, :].sum()
            Abv[n] = graph[:, n:].sum()

        measure = Cab / Aav + Cab / Abv
        sep_point = measure[1:].argmin()

        plt.plot(np.arange(-200, 600, 10), measure)
        plt.title(f'{Modality} {k} {sep_point}')

        if not os.path.exists(os.path.join('plots', 'P_value_sepration_measure',)):
            os.makedirs(os.path.join('plots', 'P_value_sepration_measure',))
        plt.savefig(os.path.join('plots', 'P_value_sepration_measure',
                                 f'{Modality}_{k}_{sep_point}.png'))


        plt.show()
```
