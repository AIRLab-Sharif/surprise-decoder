```python
import scipy.io as sio
import pandas as pd
import numpy as np

import os
```

```python
ica_size = np.zeros((18, 4))

for Sub in range(18):
    for Task in range(4):
        ica = sio.loadmat(os.path.join('ICA', f'MEG_ICA_Task{Task+1:02}_Subject{Sub+1:02}.mat'))
        ica_size[Sub, Task] = ica['data']['weights'][0, 0].shape[0]
```

```python
pd.DataFrame(ica_size, index=range(1, 19), columns=range(1, 5))
```
